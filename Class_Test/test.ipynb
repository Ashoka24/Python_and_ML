Question 2 (Pandas) – Iris Dataset Summary Report
1. Load the Iris dataset using Pandas.
2. Create a new column called petal_ratio = petal_length / petal_width.
3. Answer the following:
o a) What is the average petal_ratio per species?
o b) Which species has the highest standard deviation in sepal_length?
o c) Filter the dataset to only include rows where sepal_width is above the overall 
mean.
o d) Split the DataFrame into two subsets:
▪ One with petal_ratio < 2
▪ One with petal_ratio ≥ 2
▪ Then stack them vertically using pd.concat.

```python
#question 2
#Load the Iris dataset
df = pd.read_excel(r"C:\Users\ashok\OneDrive\Desktop\Ashoka\Python_class_Sure\Class_Test\iris.xlsx")  # or the full path if needed
df
```

```python
#Create a new column called petal_ratio = petal_length / petal_width.
df['petal_ratio'] = df['petal.length'] / df['petal.width']  #create a new column petal_ratio
df
```

```python
#What is the average petal_ratio per species?
avg_petal_ratio_per_species = df.groupby('variety')['petal_ratio'].mean() #avaerage petal_ratio per species
print("Average petal_ratio per species:") #display the average petal_ratio per species
print(avg_petal_ratio_per_species)
```

```python
#Which species has the highest standard deviation in sepal_length?
std_sepal_length = df.groupby('variety')['sepal.length'].std()
species_highest_std = std_sepal_length.idxmax()
print("Species having highest std deviation in sepal_length:", species_highest_std)
```

```python
#Filter dataset to only include rows where sepal_width is above the overall mean
mean_sepal_width = df['sepal.width'].mean()
filtered_df = df[df['sepal.width'] > mean_sepal_width] 
print("Filtered DataFrame (sepal_width > mean):")
filtered_df
```

```python
#Split DataFrame based on petal_ratio
df_low_ratio = df[df['petal_ratio'] < 2]
df_high_ratio = df[df['petal_ratio'] >= 2]
#Stack them vertically using pd.concat
stacked_df = pd.concat([df_low_ratio, df_high_ratio], axis=0)
stacked_df
```
