```python
#mean of budget and revenue 
mean_budget = df['budget'].mean()
mean_revenue = df['revenue'].mean()
#replace value less than thousand with mean value
df['budget'] = df['budget'].apply(lambda x: mean_budget if x < 1000 else x)
df['revenue'] = df['revenue'].apply(lambda x: mean_revenue if x < 1000 else x)
df
```

```python

df['profit'] = df['revenue'] - df['budget']
df
```

2.This data has information about movies made in different languages. Which language has the highest average ROI (return on investment)?

```python
#ROI is calculated as (revenue - budget) / budget * 100
df['ROI'] = (df['profit'] / df['budget']) * 100
df
```

```python
#Which language has the highest average ROI
highest_avg_ROI = df.groupby('original_language')['ROI'].mean().idxmax() 
print(highest_avg_ROI) 
#find the average ROI for the highest average ROI language with language name
highest_avg_ROI_value = df.groupby('original_language')['ROI'].mean().loc[highest_avg_ROI]
highest_avg_ROI_value
```

Find out the unique genres of movies in this dataset.

```python
# Find out the unique genres of movies in this dataset
movie_genres = df['genres'].unique()
print(movie_genres)
genre_series = df['genres'].dropna().apply(lambda g: [i['name'] for i in ast.literal_eval(g)])
unique_genres = set([genre for sublist in genre_series for genre in sublist])
unique_genres
```

Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
#Make a table of all the producers and directors of each movie.
# Create a new DataFrame with the movie title, producer, and director
producers_directors = df[['original_title', 'crew']].copy() #
producers_directors['crew'] = producers_directors['crew'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
# Extract the producer and director from the crew list
producers_directors['producer'] = producers_directors['crew'].apply(lambda x: [member['name'] for member in x if member['job'] == 'Producer']) #
producers_directors['director'] = producers_directors['crew'].apply(lambda x: [member['name'] for member in x if member['job'] == 'Director'])
producers_directors_1 = producers_directors[['original_title', 'producer', 'director']]
producers_directors_1
```

```python
# Find the top 3 producers with the highest average ROI
producer_roi = producers_directors_1[['original_title', 'producer']].explode('producer')
producer_roi = producer_roi.merge(df[['original_title', 'ROI']], on='original_title', how='left')
top_3_producers = producer_roi.groupby('producer')['ROI'].mean().sort_values(ascending=False).head(3)
top_3_producers
```

Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.

```python
#Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.
# Create a DataFrame to count the number of movies for each actor
actor_movie_count = df['cast'].apply(lambda x: [member['name'] for member in ast.literal_eval(x)] if pd.notnull(x) else []).explode().value_counts() 
actor_movie_count = actor_movie_count.reset_index()   
actor_movie_count.columns = ['actor', 'movie_count']
actor_movie_count = actor_movie_count.sort_values(by='movie_count', ascending=False)
actor_movie_count.head(10)
```

```python
#Deep dive into the movies, genres and profits corresponding to this actor
actor_name = actor_movie_count.iloc[0]['actor']
actor_movies = df[df['cast'].apply(lambda x: actor_name in [member['name'] for member in ast.literal_eval(x)] if pd.notnull(x) else False)]
actor_movies = actor_movies[['original_title', 'genres', 'profit']]
actor_movies['genres'] = actor_movies['genres'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
actor_movies['genres'] = actor_movies['genres'].apply(lambda x: [genre['name'] for genre in x])
actor_movies = actor_movies.explode('genres')
actor_movies['genres'] = actor_movies['genres'].astype(str)
actor_movies['profit'] = actor_movies['profit'].astype(float)
actor_movies = actor_movies.groupby(['original_title', 'genres'])['profit'].sum().reset_index()
actor_movies = actor_movies.sort_values(by='profit', ascending=False)
actor_movies.head(10)
```

```python
df['actors'] = df['cast'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)] if pd.notnull(x) else [])
actor_counts = df['actors'].explode().value_counts()
top_actor = actor_counts.idxmax()
print(top_actor)

# Movies featuring the top actor
top_actor_movies = df[df['actors'].apply(lambda actors: top_actor in actors)][['title', 'genres', 'profit']]
top_actor_movies
```

Top 3 directors prefer which actors the most?

```python
#Top 3 directors prefer which actors the most?
top_3_directors = producers_directors_1['director'].explode().value_counts().head(3).index.tolist()

director_actor_pref = {}
for director in top_3_directors:
    directed_movies = df[producers_directors_1['director'].apply(lambda x: director in x if isinstance(x, list) else False)]
    actors = directed_movies['actors'].explode()
    most_common_actors = actors.value_counts().head(3).index.tolist()
    director_actor_pref[director] = most_common_actors
director_actor_pref
```
